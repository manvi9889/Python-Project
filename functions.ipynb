```python
# What is a Function?

# Function ek block of code hota hai jo specific task perform karta hai.
# Jab hume same code baar-baar use karna ho, tab hum function banate hain.
# It helps in:
    # Reusability
    # Better code organization
    # Debugging easy hoti hai
    
# Types of Functions in Python:

# Built-in Functions – e.g., print(), len(), max(), sum()

# User-defined Functions – Functions created by us using def

# Lambda Functions – Small anonymous (nameless) functions

```

```python
# Syntax of User-defined Function:
# def function_name(parameters):
#     # body of function
#     return value

```

```python
# Example 1: Basic Function with Comments
# Function definition
def greet():
    print("Hello, welcome to Python!")   # this line prints greeting message

# Function call
greet()   
greet() 
greet() 
greet() # calling the function
```

```python
# Example 2: Function with Parameters and Return Value
# Function definition
def add_numbers(a, b):                  # function with two parameters
    result = a + b                      # adds both numbers
    return result    # returns the result to caller

# Function call
sum_value = add_numbers(5, 3)           # passing arguments 5 and 3
print(f"Sum is : {sum_value}")             # printing returned value
```

```python
# Example 3: Function with Default Parameter

# Function definition
def greet(name="Guest"):                # 'Guest' is default parameter value
    print("Hello", name)

# Function calls
greet("Suhail")                         # passes name "Suhail"
greet()                                 # uses default "Guest"

```

```python
# Function with Return Statement

def square(num):
    return num * num                      # returns square value

x = square(6)
print("Square is:", x)
```

```python
# Function with Multiple Return Values

def calc(a, b):
    add = a + b
    sub = a - b
    return add, sub                       # returns multiple values

x, y = calc(10, 5)
print("Addition:", x)
print("Subtraction:", y)
```

```python
# Positional and Keyword Arguments

def student(name, age):
    print("Name:", name)
    print("Age:", age)

student("Suhail", 24)                     # positional arguments
student(age=21, name="Malik")             # keyword arguments
```

```python
# Lambda (Anonymous) Functions

# Short, one-line functions — generally used for small tasks.

square = lambda x: x * x                  # lambda function
print(square(5))
```

```python
# Recursive Function

# Function that calls itself again and again until a condition is met.

def factorial(n):
    if n == 1:                            # base case
        return 1
    else:
        return n * factorial(n-1)         # recursive call

print("Factorial of 5 is:", factorial(5))


```

```python
# Local and Global Variables

x = 10                                    # global variable

def show():
    y = 5                                 # local variable
    print("Inside function:", x + y)

show()
print("Outside function:", x)
```

```python
# Passing List or Dictionary to Function

def print_list(items):
    for i in items:
        print(i)

fruits = ["apple", "banana", "cherry"]
print_list(fruits)
```

```python
# Nested Functions

def outer():
    print("Outer function called")

    def inner():
        print("Inner function called")

    inner()                               # inner function called inside outer

outer()
```

```python
# Map, Filter, Reduce with Functions


# map()
# map() ek function hai jo kisi bhi function ko har element par apply karta hai jo bhi iterable (jaise list) di ho.

# Iska syntax hai: map(function, iterable)

# Ye us function ko har item par lagaata hai, aur naya result deta hai.



# filter()
# filter() function iterable ke har element par function lagata hai aur sirf wohi items deta hai jinke liye function True return kare.

# Iska syntax hai: filter(function, iterable)

# Useful hai jab aapko kisi list mein se specific elements nikalne ho.


# reduce()
# reduce() function sab values ko cumulate kar deta hai aur ek hi value me convert kar deta hai. Ye functools module se import karna padta hai.

# Iska syntax hai: reduce(function, iterable)

# Mostly use hota hai aggregation mein, jaise sum ya product nikalna.



from functools import reduce

# map example: double all numbers
nums = [1, 2, 3, 4]
doubled = list(map(lambda x: x*2, nums))
print("Doubled:", doubled)

# filter example: even numbers
even = list(filter(lambda x: x%2==0, nums))
print("Even:", even)

# reduce example: sum of all numbers
total = reduce(lambda a,b: a+b, nums)
print("Sum:", total)
```

```python
# Problem Example

# Problem: Write a function to check whether a number is prime.

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

num = 13
if is_prime(num):
    print(num, "is Prime")
else:
    print(num, "is Not Prime")
```
